```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = plt.imread('IMG_3037.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(img, cmap='gray')
plt.show()
```

```python
img = plt.imread('IMG_3037.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blur_image = cv2.blur(img, (30, 30))
plt.axis('off')
plt.imshow(blur_image, cmap='gray')
plt.show()
```

```python
img = plt.imread('IMG_3037.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blur_image = cv2.GaussianBlur(img, (21, 21), 0)
plt.axis('off')
plt.imshow(blur_image, cmap='gray')
plt.show()
```

```python
img = plt.imread('IMG_3037.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blur_image = cv2.medianBlur(img, 5)
plt.axis('off')
plt.imshow(blur_image, cmap='gray')
plt.show()
```

```python
img = plt.imread('IMG_3037.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blur_image = cv2.bilateralFilter(img, 22, 200, 200)
plt.axis('off')
plt.imshow(blur_image, cmap='gray')
plt.show()
```

```python
def threshold(img, threshold=0.5):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            img[i,j] = 1 if img[i,j] > threshold else 0
```
